```python
from create_pysal_logo import create_logo, OUT_FILE_BASE
from create_pysal_logo import _7_NODE_COLORS, _7_NODE_NO_TEXT
import numpy

import matplotlib.pyplot as plt 
import matplotlib.image as img
%matplotlib inline


logo_file = OUT_FILE_BASE + ".png"
```

```python
alphas = numpy.linspace(0,100,7)
_7_NODE_COLORS = ["cyan!%s!magenta" % alpha for alpha in alphas]
```

```python
_7_NODE_COLORS
```

```python
"cyan!66.66666666666667!magenta", "cyan!16.666666666666668!magenta"
```

```python
n_colors = len(_7_NODE_COLORS)
color_array = numpy.empty([n_colors, n_colors], dtype=list)
```

```python
rows = color_array.shape[0]
cols = color_array.shape[1]
for i in range(rows):
    for j in range(cols):
        color_array[i,j] = _7_NODE_COLORS[i:] + _7_NODE_COLORS[:i]
```

```python
image_array = []
for i in range(rows):
    image_array.append([])
    for j in range(cols):
        # pack child node information
        node_info = numpy.array(list(zip(color_array[i,j], _7_NODE_NO_TEXT)))
        # create logo files
        create_logo(OUT_FILE_BASE, node_info)
        image_array[i].append(img.imread(logo_file))
```

```python
fig, axs = plt.subplots(rows, cols, figsize=(15, 15))
for i in range(rows):
    for j in range(cols):
        axs[i,j].imshow(image_array[i][j], interpolation="bilinear")
        axs[i,j].axis("off");
```
